```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
df = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
```

```python
%matplotlib inline
labels = pd.read_csv("validation.csv", delimiter = ";")
```

```python
data_train = df.iloc[:,1:]
data_test= test.iloc[:,1:]
data_labels = labels
```

```python
PM2 = data_train.filter(regex='PM2')
PM10 = data_train.filter(regex='PM10')
O3 = data_train.filter(regex= "O3")
NO2= data_train.filter(regex= "NO2")
PM2_test = data_test.filter(regex='PM2')
PM10_test  = data_test.filter(regex='PM10')
O3_test  = data_test.filter(regex= "O3")
NO2_test = data_test.filter(regex= "NO2")
```

```python
import seaborn as sb
sb.set_context

columns = PM2.columns
s = PM2.iloc[3]

for i in range(4):
    print columns[i*25-1]

for i in range(4):
    j = i+1
    cols = columns[i*25:j*25]
    
    vals = s[cols]
    vals.plot()
    

```

```python
targets = [PM2, PM10, O3, NO2] 
targets_test = [PM2_test, PM10_test, O3_test, NO2_test]
for target in targets:
    columns = target.columns
    idx = len(columns)/25
    for i in range(idx):
        data_train["var_%s" %columns[i*25-1]] = target[columns[i*25-1]]-target[columns[i*25-2]]
for target in targets_test:
    columns = target.columns
    idx = len(columns)/25
    for i in range(idx):
        data_test["var_%s" %columns[i*25-1]] = target[columns[i*25-1]]-target[columns[i*25-2]]

        
```

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data_train, data_labels, test_size=0.2, random_state=42)
```

```python
PM2_y = y_train.filter(regex='PM2')
PM10_y = y_train.filter(regex='PM10')
O3_y = y_train.filter(regex= "O3")
NO2_y= y_train.filter(regex= "NO2")
```

```python
import seaborn as sb
sb.set_context
from sklearn.ensemble.gradient_boosting import GradientBoostingRegressor
#Let's work on the first column of each pollutant 
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

from time import time



i=0
s = [PM2_y, PM10_y, O3_y,NO2_y]
clfs = []
for target in s:
    
    l = len(target.columns)
    col = target.columns[0]

    print col
    
    clf=GradientBoostingRegressor(n_estimators = 200)
    
    
    start = time()
    clf.fit(X_train,target[col])

    ordering = np.argsort(clf.feature_importances_)[::-1][:30]
    feature_names= X_train.columns[ordering]
    importances = clf.feature_importances_[ordering]
    x = np.arange(len(feature_names))
    plt.barh(x, importances)
    plt.yticks(x + 0.5, feature_names, fontsize=10)
    plt.show()
    #if i%24 ==0   
    end = time() - start
    print "Fitting the model took %d seconds"%end

    pred = clf.predict(X_test)
    print "MSE : ", mean_squared_error(y_test[col],pred)
    print "R2 : ",r2_score(y_test[col],pred)
```
